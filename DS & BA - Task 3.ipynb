In this regression task we will Create the Decision Tree classifier for the given data set and visualize it graphically. The purpose of this classifier is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

## STEP-1: Importing useful libraries

```python
import numpy as np                                         #Importing Numpy library for working with multidimensional arrays
import pandas as pd                                        #Importing pandas library for data analysis
import matplotlib.pyplot as plt                            #Importing matplotlib for data visualization
import sklearn.datasets as datasets                        #this library contains efficient tools for and statistical modeling
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

## STEP-2: Reading data from given file

```python
iris=datasets.load_iris()                                           #Loading the iris dataset
y=iris.target
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)     #Forming the iris dataframe
```

```python
iris_df.head()                                                      #first 5 values in the data set
```

```python
y
```

## STEP-3: Defining the Decision Tree Algorithm

```python
dtree=DecisionTreeClassifier()                        #Defining the decision tree algorithm
dtree.fit(iris_df,y)

print('Decision Tree Classifer Created')
```

## STEP-4: Graphical visualisation of the Decision Tree classifier

```python
featName=['sepal length ( in cm)','sepal width (in cm)','petal length (in cm)','petal width (in cm)']
clsName=['Iris-Setosa','Iris-Versicolor','Iris-virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (10,6), dpi = 500)

tree.plot_tree(dtree, feature_names = featName, class_names = clsName, filled = True);
```
